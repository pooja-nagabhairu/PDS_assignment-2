```python
import pandas as pd
```

```python
# Loading the dataset
data_raw = pd.read_csv('/content/train .csv')
```

```python

# dropping the unnamed column from the dataset
data_raw.drop(data_raw.columns[0], axis=1, inplace=True)
```

```python
print("Missing values in each column before filling:")
print(data_raw.isnull().sum())
```

```python
for column in data_raw.columns:
    if data_raw[column].isnull().sum() > 0:
        if data_raw[column].dtype in ['float64', 'int64']:
            data_raw[column].fillna(data_raw[column].mean(), inplace=True)
        else:
            data_raw[column].fillna(data_raw[column].mode()[0], inplace=True)
```

```python
print("\nMissing values in each column after filling:")
print(data_raw.isnull().sum())
```

```python
print("Updated DataFrame after filling missing values:")
data_raw.head()
```

```python
data_raw.to_csv('/content/data_clean.csv', index=False)
print("Data saved to /content/data_clean.csv")
```
