```python
import pandas as pd
```

```python
data = pd.read_csv('/content/data_clean.csv')
```

```python
encoded_data = pd.get_dummies(data[['Fuel_Type', 'Transmission']], drop_first=True)
encoded_data = encoded_data.astype(int)
final_data = pd.concat([data, encoded_data], axis=1)
final_data.drop(['Fuel_Type', 'Transmission'], axis=1, inplace=True)
```

```python
print("Encoded DataFrame:")
final_data.head()
```

```python
final_data.to_csv('/content/results/data_encoded.csv', index=False)

print("Data has been encoded and saved to '/content/results/data_encoded.csv'")
```
