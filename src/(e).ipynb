```python
# 4. MUTATE() - used for creating a new column (like Car_Age from Year)
current_year = datetime.now().year
mutated_data = data.copy()
mutated_data['Car_Age'] = current_year - mutated_data['Year']
print("\nData with 'Car_Age' column:\n", )
mutated_data.head()
```

```python
# 5. ARRANGE() - used for sorting by 'Price' in descending order
arranged_data = data.sort_values(by='Price', ascending=False)
print("\nData arranged by 'Price' (descending):\n", )
arranged_data.head()
```

```python
# 6. SUMMARIZE with GROUP BY - Grouping by 'Fuel_Type' and summarizing average 'Price' and 'Mileage'
summary_data = data.groupby('Fuel_Type').agg(
    Avg_Price=('Price', 'mean'),
    Avg_Mileage=('Mileage', 'mean')
).reset_index()
print("\nSummary of Average Price and Mileage by Fuel Type:\n", )
summary_data.head()
```
